subplot() function adds subplot to a current figure at the specified grid position. subplots() provides a way to plot multiple plots on a single figure. 


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x =np.array([0,1,2,3,4,5]) 
y = np.array([0,100,200,300,400,500])
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('A Line1 plot')

# for line 2 : x =np.array([0,1,2,3,4,5]) and y = 

# for line 3 : x =np.array([0,1,2,3,4,5]) and y = 

# for line 4 : x =np.array([0,1,2,3,4,5]) and y = np.array([200,350,250,550,450,150])

```

Q4. What is bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple","Microsoft","Google","AMD"])

profit = np.array([3000,8000,1000,10000])

Ans: Bar plots are a type of data visualization used to represent data in the form of rectangular bars. The height of each bar represents the value of a data point, 
and the width of each bar represents the category of the data. 

Bar charts should be used when we are showing segments of information. Vertical bar charts are useful to compare different categorical or discrete variables, 
such as age groups, classes, schools, etc., as long as there are not too many categories to compare.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple","Microsoft","Google","AMD"])
profit = np.array([3000,8000,1000,10000])
plt.barh(company,profit)
plt.xlabel("profit")
plt.ylabel("company")
plt.title("bar plot")
plt.show()
```

Q5. What is a box plot?Why is it used?Using the following data plot a box plot.

box1 = np.random.normal(100,10,200)

box2 = np.random.normal(90,20,200)

Ans : boxplot mainly provides a graphical summary of a data set with features such as minimum, first quartile, median, third quartile, and maximum.

```python
import numpy as np
import matplotlib.pyplot as plot
box1 = np.random.normal(100,10,200)
box2 = np.random.normal(90,20,200)

data = [box1,box2] 
 
fig = plot.figure(figsize =(8, 5)) 
 
# Making an axis instance
ax = fig.add_axes([0, 0, 1, 1]) 
 
# Making a plot 
bp = ax.boxplot(data,patch_artist=True) 

# display the plot 
plot.show() 
```
